```python
import pandas as pd
import numpy as np
import re
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
dataFrame = pd.read_csv("financial-Data.csv")
```

```python
dataFrame.head(20)
```

```python
dataFrame.isnull().sum()
```

```python
def metin_duzen(text):
    text = str(text)
    text = text.lower()
    text = re.sub(r"\W", " ", text)
    text = re.sub(r"\d", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
x = dataFrame["Sentence"].apply(metin_duzen)
y = dataFrame["Sentiment"].apply(metin_duzen)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
vectorizer = TfidfVectorizer(stop_words = "english", ngram_range = (1 , 2),min_df=5, max_df=0.8)
```

```python
x_train = vectorizer.fit_transform(x_train)
x_test = vectorizer.transform(x_test)
```

```python
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
# 8. Sonuçların Değerlendirilmesi
print("Doğruluk Oranı: ", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
